```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
```

```python
# Generate synthetic data (distance of riders)
np.random.seed(0)
x1 = np.random.uniform(1, 10, 50)  # Rider A's distance
x2 = np.random.uniform(1, 10, 50)  # Rider B's distance
X = np.column_stack((x1, x2))
```

```python
# Assign labels (Rider A gets order if x1 < x2, else Rider B gets it)
y = (x1 < x2).astype(int)  
```

```python
# Train a perceptron model
perceptron = Perceptron(max_iter=1000, tol=1e-3)
perceptron.fit(X, y)
```

```python
# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(0, 11, 100), np.linspace(0, 11, 100))
Z = perceptron.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
plt.scatter(x1[y == 1], x2[y == 1], color='blue', label="Rider A gets order (1)")
plt.scatter(x1[y == 0], x2[y == 0], color='red', label="Rider B gets order (0)")
plt.xlabel("Rider A's Distance")
plt.ylabel("Rider B's Distance")
plt.legend()
plt.title("Perceptron Decision Region for Order Assignment")
plt.show()
```
